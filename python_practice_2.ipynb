1.Write a Python program to accept a string value from the user and display
the count of each character in that string.

```python
string=input("enter a string value : ")
count={}
for i in string:
  if i in count:
    count[i]+=1
  else:
    count[i]=1
print(count)
```

2.  Write a Python function to find the maximum of three numbers

```python
def maximum(num1,num2,num3):
  if num1>num2 and num1>num3:
    return num1
  elif num2>num1 and num2>num3:
    return num2
  else:
    return num3
number1=int(input("enter the number 1 :"))
number2=int(input("enter the number 2 :"))
number3=int(input("enter the number 3 :"))
print(f"maximum of three numbers : {maximum(number1,number2,number3)}")
```

4. Write a Python function that takes a positive integer and returns the sum
of the cube of all the positive integers smaller than the specified number.

```python
def integer_sum(num):
  total=0
  for i in range(1,num):
    total += i**3
  return total
number=int(input("enter the number :"))
print(f"sum of numbers {integer_sum(number)}")
```

5. Write a Python program which iterates from 1 to 10. For multiples of 2,
print “Fizz” instead of the number and for the multiples of 5, print “Buzz”.
For numbers which are multiples of both 2 and 5, print “FizzBuzz”.

```python
for i in range(1,11):
  if i%2==0:
    print("Fizz",end=" ")
  elif i%5==0:
    print("buzz",end=" ")
  elif i%2 and i%5==0:
    print("FizzBuzz",end=" ")
  else:
    print(i,end=" ")
```

6. Write a Python program to find the most frequent item in a list of
numbers

```python
list1=[2, 3, 4, 2, 5, 2]
max_item=list1[0]
count=0
for item in list1:
  countmax=list1.count(item)
  if countmax>count:
    count=countmax
    max_item=item
print(f"{max_item}")

```

9. Write a Python program to convert temperatures to and from Celsius
Fahrenheit

```python
measure=input("enter Celsius or Fahrenheit ( C or F) : ").upper()
if measure=="C":
  temp=int(input("enter the temperature(Celsius) : "))
  f = (temp * 9/5) + 32
  print(f" Temperature in Fahrenheit : {f}" )
elif measure=="F":
  temp=int(input("enter the temperature(Fahrenheit) : "))
  c = (temp - 32) * 5/9
  print(f" Temperature in Celsius : {c}")
else:
  print("Invalid")
```

10. Write a Python function to calculate the factorial of a number (a
nonnegative integer). The function accepts the number as an argument.

```python
def factorial(num):
  fact=1
  for i in range(num,0,-1):
      fact*=i
  return fact
number=int(input("enter the number"))
print(factorial(number))
```
