```python
df['genre_count']=df['Genre'].str.split(',').str.len()
df['genre_count']
```

8.Handle missing values in 'Revenue (Millions)' and 'Metascore

```python
revenuemean = df['Revenue (Millions)'].mean()
metascoremean = df['Metascore'].mean()

df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(revenuemean)
df['Metascore'] = df['Metascore'].fillna(metascoremean)

print("Revenue Mean:", revenuemean)
print("Metascore Mean:", metascoremean,'\n')
print(df['Revenue (Millions)'],['Metascore'])
```

16) Create a new column Profitability:
a) Revenue divided by Duration
b) Find top 10 most profitable movies


```python
#a) Revenue divided by Duration
df['Profitability'] = df['Revenue (Millions)'] / df['Runtime (Minutes)']
df['Profitability']
```

```python
#b) Find top 10 most profitable movies
df.nlargest(10, 'Profitability')[['Title', 'Revenue (Millions)', 'Runtime (Minutes)','Profitability']]
```

17.Create a column that categorizes movies by rating: a) Rating < 6: Poor b) 6 â‰¤ Rating < 7.5: Average c) Rating â‰¥ 7.5: Excellent



```python
df["Rating_Categories"]=np.where(df["Rating"]<6,"Poor",np.where(df["Rating"]<7.5,"Average","Excellent"))
df["Rating_Categories"]
```

18.Create a custom function to score movies based on: i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score ii)Apply a function to dataset to assign the score

```python
def get_high_score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']
df['High_Score'] = df.apply(get_high_score, axis=1)
df['High_Score']
```
