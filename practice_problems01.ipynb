1. Write a Python program that takes an integer as input and calculates the
sum of its digits

```python
number =input("enter the numbers : ")
sum_of_digit=0
for digit in number:
  if digit.isdigit():
    sum_of_digit+=int(digit)
print(f"sum of digit :  {sum_of_digit}")
```

2. Create a Python program to calculate the factorial of a given number using
a function.

```python
def factorial():
  number=int(input("enter the number :"))
  fact=1
  for i in range(1,number+1):
    fact*=i
  return fact
print(f"factorial of the number is : {factorial()}")
```

3. Develop a Python program that checks whether a given word or phrase is a
palindrome.

```python
word=input("enter the word/phase :")
lower_word=word.replace(" ","").lower()
if lower_word ==lower_word[::-1]:
  print("it is a pallindrome")
else:
  print("it is not a pallindrome")
```

4. Create a Python program that takes a sentence as input and counts the
number of vowels and consonants.

```python
sentence=input("write the sentence here : ")
sentences=sentence.replace(" ","").lower()
vowels=("a","e","i","o","u")
vowels_count=0
consonants_count=0
for i in sentences:
  if i in vowels:
    vowels_count+=1

  else:

    consonants_count+=1
print(f"consonants : {consonants_count}")
print(f"vowels :{vowels_count}")

```

6. Write a Python program that takes a string as input and counts the
frequency of each character in the string.

```python
string=input("write the string : ")
count={}
for i in string:
  if i in count:
    count[i]+=1
  else:
    count[i]=1
print(count)
```

7. Develop a Python program that checks if a given year is a leap year.

```python
year=int(input("enter the year : "))
if (year%4 ==0 and year%10 != 0) or year%400==0 :
  print("it is a leap year")
else:
  print("It is not a leap year")
```
