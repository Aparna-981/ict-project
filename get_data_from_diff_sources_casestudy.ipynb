```python
import pandas as pd
import numpy as np
import requests
```

**STEP 1**:  Load SpaceX Launch Data from API

```python
url=("https://api.spacexdata.com/v4/launches")
response=requests.get(url)
launch=response.json()
launch_df = pd.DataFrame(launch)
launch_df
```

```python
launch_df.info()
```

```python
launch_df.isnull().sum()
```

```python
launch_table=launch_df[['name', 'date_utc', 'success', 'details', 'rocket']].copy()
launch_table
```

Extract relevant columns: name, date_utc, success, details, rocket


```python
launch_table['datetime'] = pd.to_datetime(launch_table['date_utc'])
launch_table['year'] = launch_table['datetime'].dt.year
```

```python
launch_table
```

Convert date_utc to datetime and extract the year


```python
launch_table_exact=launch_table.drop(columns=['date_utc'])
launch_table_exact
```

Step 2: Load Rocket Metadata


```python
rocket=pd.read_json("https://api.spacexdata.com/v4/rockets")
rocket
```

```python
rocket.info()
```

```python
rocket.isnull().sum()
```

Extract id, name, type, active, and stages


```python
rocket_table=rocket[['id', 'name', 'type', 'active','stages']].copy()
```

```python
rocket_table
```

Step 3: Merge Launch and Rocket Data

```python
df_rockets = rocket_table.set_index('id').add_suffix('_rocket')

merged_launch_rocket = launch_table_exact.join(df_rockets, on='rocket')
merged_launch_rocket
```

Step 4: Add Simulated Country Information


 Add a new column country and randomly assign one of these values:
['USA', 'Russia', 'India', 'China', 'France']


```python

countries = ['USA', 'Russia', 'India', 'China', 'France']
merged_launch_rocket['country'] = np.random.choice(countries,size=len(merged_launch_rocket))
merged_launch_rocket
```

Step 5: Store Merged Data in SQLite3
â— Use sqlite3 to create a connection and save the merged DataFrame as a table named
launches
â— Table should contain all merged columns including country

```python
import sqlite3
```

```python
con=sqlite3.connect("SpaceX_launch_rocket.db")
```

```python
merged_launch_rocket.to_sql("launches",con,index=False)
```

```python
launches_rocket=pd.read_sql("select * from launches",con)
launches_rocket
```

Step 6: Run SQL Queries on the Data to analyze
1. Launches by Country
2. Which year had the highest number of launches?
3. Top 5 Missions by Launch Count

```python
#Launches by Country
cursor=con.cursor()
cursor.execute("select count(*) as count_launches ,country from launches group by country order by count_launches desc  ")
row=cursor.fetchall()
for i in row:
  print(i)
```

```python
#Which year had the highest number of launches?
cursor.execute("select year,count(*) as count_launches from launches group by year order by count_launches   desc limit 1")
row=cursor.fetchall()
for i in row:
  print(i)
```

```python
#Top 5 Missions by Launch Count
mission ="select name as mission_count, count(*) as count_launches from launches group by name order by count_launches   desc limit 5"
pd.read_sql_query(mission,con)
```

```python
mission_rocket ="select name as mission_count, count(*) as count_launches from launches group by name_rocket order by count_launches   desc"
pd.read_sql_query(mission_rocket,con)
```
